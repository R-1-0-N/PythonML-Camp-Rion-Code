```python
# Import necessary libraries
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read synthetic data from CSV file
data = pd.read_csv('data/lab_synthetic_data.csv')
X = data[['x']].to_numpy()
y = data['y'].to_numpy()
```

```python
# Plot the data
plt.figure(figsize=(5, 5))
plt.scatter(X, y, color='blue', label='Data points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Synthetic Data')
plt.legend()
plt.show()
```

```python
model = LinearRegression()
```

```python
# fit model to data
model.fit(X, y)
```

```python
# retrieve estimated parameters, print slope and intercept
a_estimated = model.coef_[0]
b_estimated = model.intercept_
print(f"Estimated slope: {a_estimated:.2f}")
print(f"Estimated intercept: {b_estimated:.2f}")
```

```python
# generate predictions
y_pred = model.predict(X)
```

```python
# plot data and fitted line
plt.figure(figsize=(5, 5))
plt.scatter(X, y, color='blue', label='Data points')
# plot fitted line
plt.plot(X, y_pred, color='red', label='Fitted line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Fit')
plt.legend()
plt.show()
```

```python
# Calculate and print the R-squared value
r_squared = model.score(X, y)
print(f"R-squared: {r_squared:.2f}")
```
